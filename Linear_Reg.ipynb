Importing Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

Load the data

```python
df = pd.read_csv('/content/Student_Performance.csv')
```

```python
# Visualising the data
df.head()
```

```python
#check if there are null values
df.isnull().sum()
```

```python
#check the distribution of the categorical data
print(df['Extracurricular Activities'].value_counts())
```

Encoding Categorical Data

```python
#Encode the Extracurricular Activities column

df.replace({'Extracurricular Activities' : {'No' : 0, 'Yes' : 1}}, inplace=True)
```

```python
df.head()
```

Split the data into dependent and independent variables

```python
X = df.drop(['Performance Index'], axis = 1)

Y = df['Performance Index']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into testing and training

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state=2)
```

Model Training

Linear Regression

```python
#loading our linear regression model
lin_reg_model = LinearRegression()
```

```python
#fit the train data into the regression model
lin_reg_model.fit(X_train, Y_train)
```

Model Evaluation

```python
#Make prediction using the model
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
#R squared Error
error_value = metrics.r2_score(Y_train, training_data_prediction)
print("R Squared Error = ", error_value)
```

```python
#Make prediction on the test data
training_data_prediction = lin_reg_model.predict(X_test)
```

```python
#R squared Error on test data
error_value = metrics.r2_score(Y_test, training_data_prediction)
print("R Squared Error = ", error_value)
```
